```python
import pandas as pd

import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

import cleaning_functions as clean
```

```python
data = pd.read_csv('Code_challenge_train.csv')
```

```python
data = clean.cleaning(data)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('y', axis = 1)
y = data['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 72019)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
pipe = Pipeline([
    ('ss', StandardScaler()),
    ('kn', KNeighborsClassifier())
])
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe.score(X_test, y_test)
```

```python
params = {
    "kn__n_neighbors" : [5, 7, 12],
    'kn__weights' : ['uniform', 'distance']
}
```

```python
gs = GridSearchCV(estimator = pipe, param_grid = params, cv = 5)
```

```python
gs.fit(X_train, y_train)
```

```python
gs.score(X_test, y_test)
```

```python
gs.best_params_
```

As it turned out, all this time grid seaching was wasted. The best parameters are the defaults parameters for the KNN

```python
# Evaluation
#imports  
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix, f1_score
  
#---------
# predictions: classes and probabitlies 
preds = gs.predict(X_test)
preds_probs = gs.predict_proba(X_test)[:,1] # preds for positive class only
```

```python
#----------------------
# confusion matrix
cm = pd.DataFrame(confusion_matrix(y_test, preds), columns = ['Predicted 0', 'Predicted 1'], 
            index = ['Actual 0', 'Actual 1'])
cm
```

```python
#------------------------
# ROC-curve
fpr, tpr, _ = roc_curve(y_test, preds_probs)

plt.figure(figsize = (6,6))
plt.plot(fpr, tpr);
plt.plot([0,max(y_test)],[0, max(y_test)], '--'); # it takes only encoded numerical y
plt.title('ROC curve for the first model: SVM classifier'.title());
plt.xlabel('false positive rate'.title());
plt.ylabel('true positive rate'.title());

#--------------------------
# AUC scores
print(f"ROC-AUC score: {roc_auc_score(y_test, preds).round(3)}")
  
#-------------------------
# f-1 score
print(f"f1-score: {f1_score(y_test, preds).round(3)}")
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.metrics import roc_auc_score, roc_curve, f1_score
```

```python
pipe = Pipeline([
    ('ss', StandardScaler()),
    ('svc', SVC(class_weight = 'balanced', probability = True))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe.score(X_test, y_test)
```

```python
preds = pipe.predict(X_test)
preds_probs = pipe.predict_proba(X_test)[:,1]
```

```python
print(f"roc_auc score: {roc_auc_score(y_test, preds).round(3)}")
print(f"f1-score: {f1_score(y_test, preds).round(3)}")
```

```python
# ROC-curve
fpr, tpr, _ = roc_curve(y_test, preds_probs)

plt.figure(figsize = (6,6))
plt.plot(fpr, tpr);
plt.plot([0,max(y_test)],[0, max(y_test)], '--'); # it takes only encoded numerical y
plt.title('ROC curve for the SVM classifier'.title(), fontsize = 16);
plt.xlabel('false positive rate'.title(), fontsize = 14, color = 'purple');
plt.ylabel('true positive rate'.title(), fontsize = 14, color = 'purple');
```
